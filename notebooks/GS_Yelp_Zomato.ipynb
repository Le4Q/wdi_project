```python
import pandas as pd
```

```python
cor_df = pd.read_csv('../identity-resolution/data/output/gs_yelp_schema_correspondences_detail.csv', 
                     names=['name1', 'name2', 'street1','street2','postal1','postal2','id1','id2','score']).sort_values('score')
```

```python
cor_df.head()
```

Get some with very high similarity...

```python
top_sel = cor_df.loc[(cor_df['score']>0.75)]
print(len(top_sel))
top_sel_sample = top_sel.sample(95)
top_sel_sample.to_csv('../data/95_topsim_075_1.csv', sep=";")
top_sel_sample.head()
```

... check manually for correctness & corner cases

... get potential corner cases

```python
corn_sel = cor_df.loc[(cor_df['score']>0.68) & (cor_df['score']<0.8)]
print(len(corn_sel))
corn_sel.to_csv('../data/corner_cases.csv', sep=";")
corn_sel.tail()
```

... check manually for corner cases

... get negative examples

```python
neg_sel = cor_df.loc[(cor_df['score']<0.2)]
print(len(neg_sel))
neg_sel_sample = neg_sel.sample(200)
neg_sel_sample.to_csv('../data/200_lowsim_0_02.csv', sep=";")
neg_sel_sample.head()
```

Put everything together...

```python
pos = pd.read_csv('../data/95_topsim_075_1.csv', sep=";").drop('Unnamed: 0',1)
print(len(pos))
neg = pd.read_csv('../data/200_lowsim_0_02.csv', sep=";").drop('Unnamed: 0',1)
print(len(neg))
corner_pos = pd.read_csv('../data/corner_cases_pos.csv', sep=";").drop('Unnamed: 0',1)
print(len(corner_pos))
corner_neg = pd.read_csv('../data/corner_cases_neg.csv', sep=";").drop('Unnamed: 0', 1)
print(len(corner_neg))
```

```python
pos = pos.assign(match = "TRUE")
pos.head(2)
```

```python
neg = neg.assign(match = "FALSE")
neg.head(2)
```

```python
corner_pos = corner_pos.assign(match = "TRUE")
corner_pos.head(5)
```

```python
corner_neg = corner_neg.assign(match="FALSE")
corner_neg.head(2)
```

```python
total = pd.concat([pos,neg,corner_pos,corner_neg])
```

```python
total.sample(len(total))
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(total.sample(len(total))[['id1','id2','match']], test_size=0.2)
train.to_csv('../data/yelp_schema_gs_training.txt',sep=',',header=False, index=False)
test.to_csv('../data/yelp_schema_gs_test.txt',sep=',',header=False, index=False)
total.sample(len(total))[['id1','id2','match']].to_csv('../data/yelp_schema_gs_total.txt',sep=',',header=False, index=False)
```
